```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.size']=16
```

```python
def plot_barplot(timepoint):
    inputPath='/mnt/ruiyanhou/nfs_share2/three_primer/mouse_forelimb/GO/muscle_'+timepoint+'.csv'
    df=pd.read_csv(inputPath)

    selectdf=df[['Description','Log10(P)']]
    selectdf['-Log10P']=-selectdf['Log10(P)']
    selectdf
    
    importanceSelectdf=selectdf.iloc[:10,:]
    importanceSelectdf
    
    X=importanceSelectdf['Description']
    y=importanceSelectdf['-Log10P']
    
    # 数组反转。
    X=X.tolist()
    X.reverse()
    
    y=y.tolist()
    y.reverse()

    fig, ax = plt.subplots()
    b = ax.barh(range(len(X)), y, color='olive')
    
    # # 为横向水平的柱图右侧添加数据标签。
    # for rect in b:
    #     w = rect.get_width()
    #     ax.text(w, rect.get_y() + rect.get_height() / 2, '%.2f' %w, ha='left', va='center')
    
    # 设置Y轴纵坐标上的刻度线标签。
    ax.set_yticks(range(len(X)))
    ax.set_yticklabels(X)
    
    # 不要X横坐标上的label标签。
    plt.xticks(fontsize=11)
    plt.xlabel("-Log10(P value)")
    
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
    ax.spines.left.set_visible(False)

    title='Go analysis in E'+timepoint
    plt.title(title,fontweight='medium', fontsize=15,)

    # outputPath='/mnt/ruiyanhou/nfs_share2/three_primer/mouse_forelimb/GO/muscle_'+timepoint+'.pdf'
    # fig.savefig(outputPath,dpi=300,bbox_inches='tight')
    plt.show()
    

    
```

```python
for timepoint in ['10.5','11.0','12.0','13.0','14.0','15.0']:
    plot_barplot(timepoint)
    
```
