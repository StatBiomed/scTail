```python
from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor
from hilearn import CrossValidation, corr_plot
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
```

```python
#define the model objects
# linreg=linear_model.LinearRegression()
# lassoreg=linear_model.LassoCV(cv=3)
randforest=RandomForestRegressor(n_estimators=100,n_jobs=-1,random_state=666)
```

```python
from scipy.stats import pearsonr
```

```python
onegenedf=psidf[['ENSG00000143369.15']]
onegenedf
```

```python
rbpdf.drop(labels=['ENSG00000143369.15'],inplace=True,axis=1)
rbpdf
```

```python
alldf=pd.concat([rbpdf,onegenedf],axis=1)
alldf
```

```python
X=alldf.iloc[:,:-1]
X
```

```python
Y=alldf.iloc[:,-1]
Y
```

```python
X=X.values
Y=Y.values

print(X.shape)
print(Y.shape)
    
```

```python
CV = CrossValidation(X,Y)
Y_pre = CV.cv_regression(randforest)
```

```python
plt.rcParams['font.size']=16

fig=plt.subplots()
corr_plot(Y, Y_pre, size=20)
pl.xlabel("observation")
pl.ylabel("prediction")
plt.title('ECM1')
#plt.savefig('/mnt/ruiyanhou/nfs_share2/three_primer/ESCC/ERCC_part/RBP/RF_prediction/ECM1.pdf',dpi=300,bbox_inches='tight')
pl.show()
```
