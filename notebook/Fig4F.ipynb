# ZFAS1 total expression

```python
import pandas as pd
import scanpy as sc
```

```python
adata=sc.read('/mnt/ruiyanhou/nfs_share2/three_primer/human_intestinal/human_intestinal_gene_exp_scale.h5ad')
adata
```

```python
adata.X
```

```python
onegene=adata[:,adata.var.index=='ZFAS1']
onegene
```

```python
expdf=pd.DataFrame(onegene.X.toarray(),index=onegene.obs.index,columns=onegene.var.index)
expdf
```

```python
tsnedf=pd.DataFrame(onegene.obsm['X_tsne'],index=onegene.obs.index,columns=['tsne1','tsne2'])
tsnedf
```

```python
alldf=pd.concat([expdf,tsnedf],axis=1)
alldf
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 16

fig,ax=plt.subplots(figsize=(7,5))
sc=plt.scatter(alldf['tsne1'], alldf['tsne2'], c=alldf['ZFAS1'], cmap='YlOrRd',s=0.5)
position=fig.add_axes([0.82,0.56,0.03,0.33])
plt.colorbar(sc,cax=position)
#plt.set_cmap('RdBu_r')
#plt.colorbar(shrink=0.6,pad=0.01)
ax.set_xlabel('tsne_1')
ax.set_ylabel('tsne_2')
# plt.xticks("")
# plt.yticks("")
ax.set_title('ZFAS1 total expression')
# ax.set_xticks([])
# ax.set_yticks([])


# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)



outputpath='/mnt/ruiyanhou/nfs_share2/three_primer/human_intestinal/figure/example/ZFAS1_exp.pdf'
#fig.savefig(outputpath,dpi=300,bbox_inches='tight')
plt.show()
```
