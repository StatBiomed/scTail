# get the percentage of A T C G in each position

```python
import pandas as pd
import matplotlib.pyplot as plt

from Bio import SeqIO
import numpy as np
```

```python
matrix_col=['pos'+str(i) for i in range(-100,101)]
matrix_col
```

```python
num_matrix=pd.DataFrame(columns=matrix_col,index=['A','T','C','G'])
num_matrix
```

```python
matrixdf=num_matrix.fillna(0)
matrixdf
```

```python
for pos in range(0,201):

    for record in SeqIO.parse("/mnt/ruiyanhou/nfs_share2/three_primer/technique_compare/ATCG_distribution/k562_sierra_PAS_up_down_100.fa","fasta"):

        nucleotide=record.seq[pos]
        if nucleotide=='A':
            matrixdf.iloc[0,pos]+=1

        elif nucleotide=='T':
            matrixdf.iloc[1,pos]+=1

        elif nucleotide=='C':
            matrixdf.iloc[2,pos]+=1

        elif nucleotide=='G':
            matrixdf.iloc[3,pos]+=1
            
```

```python
matrixdf
```

```python
matrixdfcopy=matrixdf.copy()
matrixdfcopy
```

```python
matrixdfcopyT=matrixdfcopy.T
matrixdfcopyT
```

```python
matrixdfcopyT['sum']=matrixdfcopyT.sum(axis=1)
matrixdfcopyT
```

# plot

```python
import pandas as pd
import matplotlib.pyplot as plt

from Bio import SeqIO
import numpy as np
```

```python
matrixdfcopyT=pd.read_csv('/mnt/ruiyanhou/nfs_share2/three_primer/technique_compare/ATCG_distribution/k562_sierra_PAS_up_down_100_matrix.csv',index_col=0)
matrixdfcopyT
```

```python
matrixdfcopyT=matrixdfcopyT.apply(lambda x:x/matrixdfcopyT['sum'])
matrixdfcopyT
```

```python
matrixdfcopyT['true_pos']=[i for i in range(-100,101)]
matrixdfcopyT
```

```python
fig, ax = plt.subplots(figsize=(7,5))
plt.rcParams['font.size'] = 16

ax.plot(matrixdfcopyT['true_pos'], matrixdfcopyT['A'],label='A',color='g')
ax.plot(matrixdfcopyT['true_pos'], matrixdfcopyT['T'],label='T',color='r')
ax.plot(matrixdfcopyT['true_pos'], matrixdfcopyT['C'],label='C',color='b')
ax.plot(matrixdfcopyT['true_pos'], matrixdfcopyT['G'],label='G',color='orange')
ax.axvline(x=0, color='black', linestyle='--')

ax.set(xlabel='Relative position to polyA cleavage site', ylabel='Nucleotide ratio',
       title='PAS detected by Sierra')
#ax.grid()
plt.legend()
#fig.savefig("/mnt/ruiyanhou/nfs_share2/three_primer/technique_compare/ATCG_distribution/k562_sierra_pas.pdf",dpi=300,bbox_inches='tight')
plt.show()
```
