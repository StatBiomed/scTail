```python
import pandas as pd
import scanpy as sc
```

```python
allls=[]
for cell in ['Enteriendocrine','Enterocyte','Goblet','Paneth-like','Progenitor','Stem_Cell','TA']:
    inputfile='/mnt/ruiyanhou/nfs_share2/three_primer/human_intestinal/run_BRIE/Brie_out/brie_'+cell+'.brie_ident.tsv'
    alldf=pd.read_csv(inputfile,delimiter='\t')
    celltype=cell.replace('_',' ')
    col_name='celltype_'+celltype+'_FDR'
    signdf=alldf[alldf[col_name]<0.05]
    genels=signdf['GeneID'].tolist()
    print(genels)
    allls.append(genels)
```

```python
flattenallls=[j for i in allls for j in i]
flattenallls
```

```python
flattenset=set(flattenallls)
flattenset
```
