```python
import pandas as pd
```

```python
countdf=pd.DataFrame({'type':['annotation','unannotation'],'number':[14660,5216]})
countdf
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.rcParams['legend.handlelength'] = 1
matplotlib.rcParams['legend.handleheight'] = 1
```

```python
fig, ax = plt.subplots(figsize=(6, 3))



#attentaion: ax.pie在countdf['feature'].values也会返回一些值给autopct,所以absolute_value需要一个val参数，这个val指的就是countdf['feature'].values
# def absolute_value(val):
#     a  = '{:.0f}'.format(val/100*countdf['number'].values.sum())
#     return a


ax.pie(countdf['number'].values,colors=['lightcoral','darkturquoise'],autopct='%1.1f%%')



ax.legend(countdf['type'],
          loc="center left",
          bbox_to_anchor=(0.94, 0, 0.5, 1),frameon=False)

#fig.savefig("/mnt/ruiyanhou/nfs_share2/three_primer/technique_compare/annotation_percent/K562_pct_unannotaion_annotation.pdf", dpi=300, bbox_inches='tight')


plt.show()
```
