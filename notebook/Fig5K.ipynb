```python
selectdf

from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor
from hilearn import CrossValidation, corr_plot
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
```

```python
#define the model objects
# linreg=linear_model.LinearRegression()
# lassoreg=linear_model.LassoCV(cv=3)
randforest=RandomForestRegressor(n_estimators=100,n_jobs=-1,random_state=666)
```

```python
from scipy.stats import pearsonr
```

```python
pearsonR={}

for oney in selectdf.columns:
    print(oney)
    ydf=selectdf[oney]
    print(ydf)
    if oney in rbpdf.columns.tolist():
        print("hello")
        rbpdf.drop(labels=[oney],axis=1,inplace=True)
        

    
    
    totaldf=pd.concat([rbpdf,ydf],axis=1)
    #print(totaldf)
    X=totaldf.iloc[:,:-1]
    Y=totaldf[oney]

    X=X.values
    Y=Y.values

    print(X.shape)
    print(Y.shape)
    
        

    CV = CrossValidation(X,Y)
    Y_pre = CV.cv_regression(randforest)

    corr_coef, p_value = pearsonr(Y, Y_pre)
    pearsonR[oney]=[corr_coef,p_value]
    
```
