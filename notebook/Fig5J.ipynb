```python
refdf=pd.read_csv('/mnt/ruiyanhou/nfs_share2/three_primer/ESCC/run_scTail/scTail_callpeak/SRR16796861/ref_file/ref_gene.tsv',delimiter='\t')
refdf=refdf[['gene_id','gene_name']]
refdf
```

```python
#outdf.to_csv('/mnt/ruiyanhou/nfs_share2/three_primer/ESCC/ERCC_part/RBP/gap_binding_RBP_withanno.csv')

selectadata=adata[:,adata.var['gene_ids'].isin(outdf['gene_id'])]
selectadata
```

```python
selectadata.obs['celltype'].unique()
```

```python
fibrablastadata=selectadata[selectadata.obs['celltype']=='Epithelial cell',:]
fibrablastadata
```

```python
fibrablastadata.var.index

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import gcf
```

```python
alldfls=[]
for patientid in fibrablastadata.obs['PatientType'].unique():
    for sampletype in fibrablastadata.obs['SampleType'].unique():
        onepatientadata=fibrablastadata[(fibrablastadata.obs['PatientType']==patientid)&(fibrablastadata.obs['SampleType']==sampletype),:]
        rbpexp=np.mean(onepatientadata.X.toarray(),axis=0)
        coltitle=str(patientid)+'_'+sampletype
        rbpexpdf=pd.DataFrame(rbpexp,index=onepatientadata.var.index,columns=[coltitle])
    
        alldfls.append(rbpexpdf)
        
```

```python
alldfls

finaldf=pd.concat(alldfls,axis=1)
finaldf
```

```python
finaldf.dropna(how='any',inplace=True,axis=1)
finaldf
```

```python
import seaborn as sns
```

```python
columndf=pd.DataFrame(finaldf.columns,columns=['name'])
columndf
```

```python
columndf['condition']=columndf['name'].str.split('_',expand=True)[1]
columndf
```

```python
columndf.sort_values(['condition'],inplace=True)
columndf
```

```python
lut1=dict(zip(set(columndf['condition']), ['gold','lightseagreen']))
column_colors=columndf['condition'].map(lut1)

column_colors
```

```python
finaldf=finaldf[columndf['name'].tolist()]
finaldf
```

```python
finaldf.reset_index(inplace=True)
finaldf
```

```python
finaldf.rename(columns={'index':'gene_name'},inplace=True)
finaldf
```

```python
outdf=outdf[['RBP','gene_name']]
outdf
```

```python
outdf.drop_duplicates(['RBP','gene_name'],inplace=True)
outdf
```

```python
len(outdf)
```

```python
finaldf=outdf.merge(finaldf,on='gene_name')
finaldf
```

```python
finaldf.drop(labels='gene_name',inplace=True,axis=1)
finaldf
```

```python
finaldf.set_index('RBP',inplace=True)
finaldf
```

```python
plt.rcParams['font.size']=15
```

```python
fig=sns.clustermap(finaldf,metric='euclidean',method='average',cmap='YlGnBu',col_colors=[column_colors],col_cluster=True,yticklabels=True,row_cluster=True,
                  figsize=(6,9),colors_ratio=0.02)

#fig=sns.clustermap(fisherplotdf,metric='euclidean',method='ward',row_colors=[row_colors],col_colors=column_colors,cmap='YlGnBu',col_cluster=True)
#fig.ax_col_dendrogram.set_visible(False)
fig.ax_heatmap.set_xlabel(" ")
fig.ax_heatmap.set_ylabel(" ")





for label in columndf['condition'].unique():
    fig.ax_row_dendrogram.bar(0, 0, color=lut1[label],label=label, linewidth=0)
l2=fig.ax_row_dendrogram.legend(title='Condition', loc="center", ncol=2, bbox_to_anchor=(0.82, 0.97), bbox_transform=gcf().transFigure) 

#fig.savefig('/mnt/ruiyanhou/nfs_share2/three_primer/ESCC/ERCC_part/RBP/gap_sequence_tumor_normal_heatmap.pdf',dpi=300,bbox_inches='tight')
plt.show()
```
