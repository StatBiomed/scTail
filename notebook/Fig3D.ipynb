```python
utr5df=pd.read_csv('/mnt/ruiyanhou/nfs_share2/annotation/genomic_region_annotation/utr5.bed',delimiter='\t',header=None)
utr5df

utr5df['Feature']='UTR5'
utr5df
```

```python
utr3df=pd.read_csv('/mnt/ruiyanhou/nfs_share2/annotation/genomic_region_annotation/utr3.bed',delimiter='\t',header=None)
utr3df

utr3df['Feature']='UTR3'
utr3df
```

```python
introndf=pd.read_csv('/mnt/ruiyanhou/nfs_share2/annotation/genomic_region_annotation/intron.bed',delimiter='\t',header=None)
introndf

introndf['Feature']='intron'
introndf
```

```python
exondf=pd.read_csv('/mnt/ruiyanhou/nfs_share2/annotation/genomic_region_annotation/exon.bed',delimiter='\t',header=None)
exondf

exondf['Feature']='exon'
exondf
```

```python
alldf=pd.concat([utr5df,utr3df,introndf,exondf],axis=0)
alldf

alldf.columns=['chromosome','start','end','gene_id','score','strand','Feature']
alldf
```

```python
import scanpy as sc
import numpy as np
```

```python
adata=sc.read('/mnt/ruiyanhou/nfs_share2/three_primer/technique_compare/run_scTail/k562_scRNA_seq_mincount_50/count/all_cluster.h5ad')
adata
```

```python
pasdf=adata.var.copy()
pasdf
```

```python
pasdf['PAS']=np.where(pasdf['cluster_strand']=='+',pasdf['cluster_end'],pasdf['cluster_start'])
pasdf
```

```python
pasdf.reset_index(inplace=True)
pasdf
```

```python
pasdf=pasdf[['cluster_id','gene_id','PAS']]
pasdf
```

```python
pasdf['gene_id']=pasdf['gene_id'].str.split('.',expand=True)[0]
pasdf
```

```python
## intron

featurels=[]
for i in range(0,len(pasdf)):
    genename=pasdf['gene_id'][i]
    selectrefdf=alldf[alldf['gene_id']==genename]
    selectrefdf.reset_index(inplace=True,drop=True)
    PAS=pasdf['PAS'][i]
    name=pasdf['cluster_id'][i]
    # print(selectrefdf)
    # print(PAS)
    
    for j in range(0,len(selectrefdf)):
        start=selectrefdf['start'][j]
        end=selectrefdf['end'][j]
        reffeature=selectrefdf['Feature'][j]
        if (PAS>start)&(PAS<end):
            featurels.append([name,reffeature])
```

```python
featurels
```

```python
featureassigndf=pd.DataFrame(featurels)
featureassigndf
```

```python
featureassigndf.columns=['TSS_name','feature']
featureassigndf
```

```python
keepfirstdf=featureassigndf.drop_duplicates(['TSS_name'])
keepfirstdf
```

```python
keepfirstdf['feature'].value_counts()
```

# plot pie chart 

```python
import pandas as pd
```

```python
df=pd.read_csv('/mnt/ruiyanhou/nfs_share2/three_primer/technique_compare/genomic_distribution/k562_features_distribution.csv',index_col=0)
df
```

```python
countdf=pd.DataFrame(df['feature'].value_counts())
countdf
```

```python
countdf=countdf.reindex(['UTR5','intron','exon','UTR3'])
countdf
```

```python
countdf.index=["5'UTR","CDS-intron","CDS-exon","3'UTR"]
countdf
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.rcParams['legend.handlelength'] = 1
matplotlib.rcParams['legend.handleheight'] = 1
```

```python
fig, ax = plt.subplots(figsize=(6, 3))

#attentaion: ax.pie在countdf['feature'].values也会返回一些值给autopct,所以absolute_value需要一个val参数，这个val指的就是countdf['feature'].values
def absolute_value(val):
    a  = '{:.0f}'.format(val/100*countdf['count'].values.sum())
    return a


ax.pie(countdf['count'].values,colors=['lightseagreen','grey','forestgreen','orange'],autopct=absolute_value)

ax.legend(countdf.index,
          loc="center left",
          bbox_to_anchor=(0.94, 0, 0.5, 1),frameon=False)
#plt.title('Unannotation mode')

#fig.savefig("/mnt/ruiyanhou/nfs_share2/three_primer/technique_compare/genomic_distribution/k562_genomic_distribution.pdf", dpi=300, bbox_inches='tight')

plt.show()
```
